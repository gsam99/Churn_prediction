```python
def rare_samples_checking(data,col,min_samples=100):
    value_count_dict=data[col].value_counts().to_dict()
    cat_list=[]
    for v, i in value_count_dict.items():
        if i>=min_samples:
            cat_list.append(v)
    return cat_list

dummies = pd.get_dummies(df["channel_sales"], dtype=int)
rare_cats=rare_samples_checking(df,"channel_sales")
dummies=dummies[rare_cats]
dummies

```

```python
#df.drop(["channel_sales"],axis=1,inplace=True)
df=pd.concat([df,dummies],axis=1)
```
